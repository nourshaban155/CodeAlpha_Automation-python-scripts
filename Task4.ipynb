```python
import os
import shutil

# تحديد مجلد المصدر
source_folder = r"D:\bbc\business"

# أنواع الملفات والمجلدات الخاصة بها
file_types = {
    "Images": [".png", ".gif", ".jpg"],
    "Documents": [".pdf", ".txt", ".pptx"]
}

# إنشاء المجلدات إذا لم تكن موجودة
for folder in file_types.keys():
    folder_path = os.path.join(source_folder, folder)
    os.makedirs(folder_path, exist_ok=True)

# إنشاء مجلد "Other" للملفات غير المصنفة
other_folder = os.path.join(source_folder, "Other")
os.makedirs(other_folder, exist_ok=True)

# فرز الملفات ونقلها إلى المجلدات المناسبة
for file in os.listdir(source_folder):
    file_path = os.path.join(source_folder, file)

    # التحقق مما إذا كان العنصر ملفًا وليس مجلدًا
    if os.path.isfile(file_path):
        file_extension = os.path.splitext(file)[1].lower()
        moved = False  # متغير للتحقق مما إذا تم نقل الملف

        for folder, extensions in file_types.items():
            if file_extension in extensions:
                shutil.move(file_path, os.path.join(source_folder, folder, file))
                print(f"Moved {file} to {folder}")
                moved = True
                break

        # إذا لم يتم نقل الملف، انقله إلى "Other"
        if not moved:
            shutil.move(file_path, os.path.join(other_folder, file))
            print(f"Moved {file} to Other")

print("Folder organization completed successfully!")
```

```python
import pandas as pd

# 1️⃣ تحميل ملف الإكسل
file_path = r"D:\document\O Gad.xlsx"  # تأكد من صحة المسار
df = pd.read_excel(file_path, engine="openpyxl")

# 2️⃣ حذف الصفوف والأعمدة الفارغة بالكامل
df.dropna(how="all", inplace=True)  # حذف الصفوف الفارغة بالكامل
df.dropna(axis=1, how="all", inplace=True)  # حذف الأعمدة الفارغة بالكامل

# 3️⃣ معالجة القيم الفارغة (ملؤها بـ "NA" بدلًا من حذفها)
df.fillna("NA", inplace=True)

# 4️⃣ إزالة التكرار
df.drop_duplicates(inplace=True)

# 5️⃣ تنظيف الأعمدة النصية
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].astype(str).str.strip()  # إزالة المسافات الزائدة
    if "email" in col.lower():
        df[col] = df[col].str.lower()  # تحويل البريد الإلكتروني إلى حروف صغيرة
    else:
        df[col] = df[col].str.title()  # تنسيق الأسماء والعناوين

# 6️⃣ تحويل الأعمدة التي تحتوي على "date" إلى صيغة التاريخ
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")  # تحويل إلى تاريخ مع تجنب الأخطاء
        df[col].fillna("NA", inplace=True)  # استبدال القيم غير الصحيحة بـ "NA"

# 7️⃣ حفظ الملف الجديد بعد التنظيف
cleaned_file_path = r"D:\document\cleaned_labb5.xlsx"
df.to_excel(cleaned_file_path, index=False, engine="openpyxl")

print(f"✅ تم تنظيف البيانات وحفظها في: {cleaned_file_path}")
```

```python
import os
import psutil
import datetime
import tempfile
import shutil

def clear_temp_files():
    temp_dir = tempfile.gettempdir()
    for filename in os.listdir(temp_dir):
        file_path = os.path.join(temp_dir, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
            print(f"Deleted: {file_path}")
        except Exception as e:
            print(f"Failed to delete {file_path}: {e}")

def system_maintenance():
    uptime = datetime.datetime.now() - datetime.datetime.fromtimestamp(psutil.boot_time())
    cpu_usage = psutil.cpu_percent(interval=1)
    ram = psutil.virtual_memory()
    disk = shutil.disk_usage("/")

    print(f"Uptime: {uptime}")
    print(f"CPU: {cpu_usage}%")
    print(f"RAM: {ram.percent}% ({ram.used // (1024**3)}GB / {ram.total // (1024**3)}GB)")
    print(f"Disk: {disk.free // (1024**3)}GB free / {disk.total // (1024**3)}GB")

def optimize_memory():
    if os.name == 'nt':
        os.system('wmic process where "WorkingSetSize>1000000" call Terminate')
    else:
        os.system('sync; echo 3 > /proc/sys/vm/drop_caches')
    print("Memory optimized.")

def main():
    print("Starting system maintenance...")
    clear_temp_files()
    optimize_memory()
    system_maintenance()
    print("System maintenance completed!")

if __name__ == "__main__":  
    main()
```
